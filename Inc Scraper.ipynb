```python
#Let's see if we can format the name like it would be in a crunchbase url. Hint: we can.
name = inc5000['Company Name'][2]
splits = name.split()
word = ''
for i in range(len(splits)-1):
    word += splits[i].lower() + '-'
word += splits[-1].lower()
word = str(word)
word
```

```python
#Let's put all this in a function and test it out.
def websetup(name):
    splits = name.split()
    word = ''
    for i in range(len(splits)-1):
        word += splits[i].lower() + '-'
    word += splits[-1].lower()
    #word = str(word)
    url = 'http://www.crunchbase.com/organization/' + word
    return(url)
websetup(inc5000['Company Name'][2])
```

```python
#Let's put all these urls in a list.
crunchurls = [websetup(name) for name in inc5000['Company Name']]
```

```python
#Groovy
crunchurls[0:5]
```

```python
#Now we need a function to check which of these urls actually exists
def crunchchecker(browser,url):
    browser.get(url)
    if browser.title != no:
        return url
    else:
        return 'Page not found'
```

```python
#Let's work these functions
crunchurls = [crunchchecker(browser,url) for url in crunchurls]
```

```python
#We'll write these to another file to save them
file_object = open('crunchurls.txt','w')
for item in crunchurls:
  file_object.write("%s\n" % item)
file_object.close()
```
